```python

import matplotlib.pyplot as plt
import numpy as np
rollno=np.arange(1,11)
series1=np.array([10,11,12,13,14,15,16,12,14,12])
series2=np.array([12,15,19,18,17,14,13,12,15,16])
plt.plot(rollno,series1)
plt.plot(rollno,series2)
plt.xlabel("Rollno")
plt.ylabel("mark")
plt.plot(rollno,series1)
plt.plot(rollno,series2)
plt.subplot(1,2,1)
plt.plot(series1,color='green',linestyle='-')
plt.show()
```

```python
Series1=np.array([10,11,12,13,14,15,16,12,14,12])
Series2=np.array([12,15,19,18,17,14,13,12,15,16])
plt.scatter(rollno,series1)
plt.scatter(rollno,series2)
plt.legend(["Series1","Series2"])
plt.xlabel("rollno")
plt.ylabel("mark1")
plt.title("series1 performance")
plt.show()
```
